## Import necessary modules for work

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Create a DataFrame with prepared data

```python
df = pd.read_csv("vacancy.csv")
df
```

## Data cleaning

This step may involve data cleaning, data transformation, and feature engineering to make the data suitable for analysis.

```python
series_skills = df["skills"].str.split(",").explode()
skills = pd.get_dummies(series_skills).groupby(level=0).max().sum()
skills
```

## Visualization


```python
df["city"].value_counts().plot(kind="bar")
plt.title("Розподіл вакансій за містами")
plt.xlabel("Місто")
plt.ylabel("Кількість")
plt.show()
```

```python
top_skills = skills.sort_values(ascending=False).head(25)
top_skills.plot(kind="bar")
plt.title("ТОП 25 навичок за розподілом")
plt.xlabel("Навичка")
plt.ylabel("Частота")
plt.show()
```
